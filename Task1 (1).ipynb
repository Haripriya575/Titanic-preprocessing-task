```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load your uploaded CSV file
df = pd.read_csv("Titanic-Dataset.csv")

# Show first few rows
df.head()
```

```python
# Dataset info
df.info()

# Missing values
print("\nMissing Values:\n", df.isnull().sum())

# Summary statistics
df.describe()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df = df.drop('Cabin', axis=1, errors='ignore')
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
# Convert 'Sex' to numeric (0 = male, 1 = female)
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

# One-hot encode 'Embarked' column
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Apply scaling to Age and Fare columns
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])
```

```python
# Boxplot to visualize outliers in Fare
sns.boxplot(x=df['Fare'])
plt.title("Fare - Boxplot for Outlier Detection")
plt.show()
```

```python
# Calculate IQR
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
df = df[(df['Fare'] >= lower_bound) & (df['Fare'] <= upper_bound)]
```

```python
df.to_csv("cleaned_titanic.csv", index=False)
```
